```python
import numpy as np
m = np.matrix('1 2 3 4; 5 6 7 8; 9 1 5 7')
print(m)
```

```python
print(m[1, 0])
```

```python
print(m[1, :])
```

```python
print(m[:, 2])
```

```python
print(m[1, 2:])
```

```python
print(m[0:2, 1])
```

```python
print(m[0:2, 1:3])
```

```python
cols = [0, 1, 3]
print(m[:, cols])
```

```python
print(type(m))
```

```python
m = np.array(m)
print(type(m))
```

```python
print(m.shape)
```
