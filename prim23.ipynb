```python
import numpy as np
A = np.matrix('1 2 3; 4 5 6')
C = 3 * A
print(C)
```

```python
B = np.matrix('1 2; 3 4')
L = 1 * B
R = B
print(L)
print(R)
```

```python
Z = np.matrix('0 0; 0 0')
L2 = 0 * B
R2 = Z
print(L2)
print(R2)
```

```python
p = 2
q = 3
L3 = (p + q) * B
R3 = p * B + q * B
print(L3)
print(R3)
```

```python
L4 = (p * q) * B
R4 = p * (q * B)
print(L4)
print(R4)
```

```python
C = np.matrix('5 6; 7 8')
k = 3
L5 = k * (B + C)
R5 = k * B + k * C
print(L5)
print(R5)
```
