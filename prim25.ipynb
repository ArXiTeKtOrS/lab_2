```python
import numpy as np
A = np.matrix('1 2; 3 4')
B = np.matrix('5 6; 7 8')
C = np.matrix('2 4; 7 8')
L = A.dot(B.dot(C))
R = (A.dot(B)).dot(C)
print(L)
print(R)
```

```python
L2 = A.dot(B + C)
R2 = A.dot(B) + A.dot(C)
print(L2)
print(R2)
```

```python
L3 = A.dot(B)
R3 = B.dot(A)
print(L3)
print(R3)
```

```python
E = np.matrix('1 0; 0 1')
I = E.dot(A)
Q = A.dot(E)
print(I)
print(Q)
```

```python
Z = np.matrix('0 0; 0 0')
I2 = Z.dot(A)
Q2 = A.dot(Z)
print(I2)
print(Q2)
```
