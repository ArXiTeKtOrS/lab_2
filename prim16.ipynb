```python
import numpy as np
print(np.random.permutation(7))
a = ['a', 'b', 'c', 'd', 'e']
print(np.random.permutation(a))
```

```python
arr = np.linspace(0, 10, 5)
print(arr)
```

```python
arr_mix = np.random.permutation(arr)
print(arr_mix)
```

```python
index_mix = np.random.permutation(len(arr_mix))
print(index_mix)
print(arr[index_mix])
```
