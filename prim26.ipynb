```python
import numpy as np
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
print(A)
```

```python
print(np.linalg.det(A))
```

```python
print(A)
print(A.T)
det_A = round(np.linalg.det(A), 3)
det_A_t = round(np.linalg.det(A.T), 3)
print(det_A)
print(det_A_t)
```

```python
B = np.matrix('-4 -1 2; 0 0 0; 8 3 1')
print(B)
print(np.linalg.det(B))
```

```python
C = np.matrix('10 4 -1; -4 -1 2; 8 3 1')
print(A)
print(C)
print(round(np.linalg.det(A), 3))
print(round(np.linalg.det(C), 3))
```

```python
D = np.matrix('-4 -1 2; -4 -1 2; 8 3 1')
print(D)
print(np.linalg.det(D))
```

```python
print(A)
k = 2
E = A.copy()
E[2, :] = k * E[2, :]
print(E)
det_E = round(np.linalg.det(E), 3)
print(det_A * k)
print(det_E)
```
