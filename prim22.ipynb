```python
import numpy as np
A = np.matrix('1 2 3; 4 5 6')
print(A)
```

```python
A_t = A.transpose()
print(A_t)
```

```python
print(A.T)
```

```python
R = (A.T).T
print(R)
```

```python
B = np.matrix('7 8 9; 0 7 5')
L = (A + B).T
R2 = A.T + B.T
print(L)
print(R2)
```

```python
C = np.matrix('1 2; 3 4')
D = np.matrix('5 6; 7 8')
L2 = (C.dot(D)).T
R3 = (D.T).dot(C.T)
print(L2)
print(R3)
```

```python
k = 3
L3 = (k * A).T
R4 = k * (A.T)
print(L3)
print(R4)
```

```python
C_det = np.linalg.det(C)
C_T_det = np.linalg.det(C.T)
print(f"{C_det:.9f}")
print(f"{C_T_det:.9f}")
```
