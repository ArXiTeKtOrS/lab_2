# Задание 5
### Создайте: Единичную матрицу размером 4×4. Диагональную матрицу размером 4×4 с диагональными элементами [5, 10, 15, 20] (не использовать циклы). Найдите сумму всех элементов каждой из этих матриц и сравните результаты.

```python
ed = np.eye(4)
diag = np.diag([5, 10, 15, 20])

sum_ed = np.sum(ed)
sum_diag = np.sum(diag)

print("Сумма элементов единичной матрицы:", sum_ed)
print("Сумма элементов диагональной матрицы:", sum_diag)

if sum_ed > sum_diag:
    print("Сумма элементов единичной матрицы больше")
elif sum_ed < sum_diag:
    print("Сумма элементов диагональной матрицы больше")
else:
    print("Суммы равны")
```

# Задание 8
### Создайте случайную квадратную матрицу 3×3. Найдите и выведите: Определитель этой матрицы, Обратную матрицу (если существует, иначе выведите сообщение, что матрица вырождена). Используйте функции np.linalg.det и nр.linalg.inv.

```python
A = np.random.rand(3, 3)
print("Случайная матрица:\n", A)

A_det = round(np.linalg.det(A), 3)
print("\nОпределитель матрицы", A_det)

if A_det == 0:
    print("\nМатрица вырождена!")
else:
    A_inv = np.linalg.inv(A)
    print("\nОбратная матрица:\n", A_inv)
```

# Задание 10
### Решите систему линейных уравнений вида:
#### 2x + 3y - z = 5
#### 4x - y + 2z = 6
#### -3x + 5y + 4z = -2

```python
A = np.array([[2, 3, -1],
              [4, -1, 2],
              [-3, 5, 4]])

B = np.array([5, 6, -2])

x = np.linalg.solve(A, B)
print("Решение системы (x, y, z):", x)
```

# Индивидуальное задание
### Задание предусматривает построение системы линейных уравнений. Решите полученную систему уравнений с использованием библиотеки NumPy. Для решения системы используйте метод Крамера и матричный метод. Сравните полученные результаты, с результатами, полученными с помощью np.linalg.solve.
### Условие: Распределение инвестиций. Компания выделила 1 000 000 рублей на три проекта. Первый проект получает в два раза больше, чем второй, а третий получает на 200 000 рублей больше, чем второй. Сколько денег получит каждый проект?
### Пусть: x1 - деньги для 1 проекта, x2 - деньги для 2 проекта, x3 - деньги для 3 проекта
### Общая сумма: x1 + x2 + x3 = 1000000, x1 = 2x2, x1 - 2x2 = 0, x3 = x2 + 200000, -x2 + x3 = 200000
### Тогда получаем систему уравнений:
### x1 + x2 + x3 = 1000000
### x1 - 2x2 = 0
### -x2 + x3 = 200000

```python
import numpy as np

# Матрица коэффициентов
A = np.array([
    [1, 1, 1],
    [1, -2, 0],
    [0, -1, 1]
], dtype=float)

# Вектор правой части
B = np.array([1000000, 0, 200000], dtype=float)

print("=== Метод Крамера ===")
# Вычисляем определитель A
det_A = np.linalg.det(A)

# Создаём копии A и заменяем столбцы
def cramer(A, B, i):
    A_i = A.copy()
    A_i[:, i] = B
    return np.linalg.det(A_i) / det_A

x1 = cramer(A, B, 0)
x2 = cramer(A, B, 1)
x3 = cramer(A, B, 2)

print(f"x1 = {x1:,.0f} руб")
print(f"x2 = {x2:,.0f} руб")
print(f"x3 = {x3:,.0f} руб")


print("\n=== Матричный метод (через обратную матрицу) ===")
A_inv = np.linalg.inv(A)
X_mat = A_inv @ B
print(f"x1 = {X_mat[0]:,.0f} руб")
print(f"x2 = {X_mat[1]:,.0f} руб")
print(f"x3 = {X_mat[2]:,.0f} руб")

print("\n=== Метод np.linalg.solve ===")
X_solve = np.linalg.solve(A, B)
print(f"x1 = {X_solve[0]:,.0f} руб")
print(f"x2 = {X_solve[1]:,.0f} руб")
print(f"x3 = {X_solve[2]:,.0f} руб")

print("\n=== Проверка ===")
print("Все методы дают одинаковый результат:", np.allclose(X_mat, X_solve))
```
