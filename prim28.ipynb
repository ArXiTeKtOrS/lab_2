```python
import numpy as np
A = np.matrix('1 -3; 2 5')
A_inv = np.linalg.inv(A)
print(A_inv)
```

```python
A_inv_inv = np.linalg.inv(A_inv)
print(A)
print(A_inv_inv)
```

```python
L = np.linalg.inv(A.T)
R = (np.linalg.inv(A)).T
print(L)
print(R)
```

```python
B = np.matrix('7 6; 1 8')
L2 = np.linalg.inv(A.dot(B))
R2 = np.linalg.inv(B).dot(np.linalg.inv(A))
print(L2)
print(R2)
```
